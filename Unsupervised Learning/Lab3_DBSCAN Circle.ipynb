### Visualize data

```python
import seaborn as sns
%matplotlib inline
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.scatterplot(x='Xcircle_X1', y='Xcircle_X2', data=df, palette='spring')
```

### Applying K-Means

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=2)
df['KM_Cluster'] = km.fit_predict(df[['Xcircle_X1', 'Xcircle_X2']])
```

Visualize K-Means cluster

```python
sns.scatterplot(x='Xcircle_X1', y='Xcircle_X2', hue='KM_Cluster', data=df, palette='spring')
```

### Applying DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
#Initialize DBSCAN
db = DBSCAN(eps=0.2, min_samples=10)
```

```python
#Fit on the data
db.fit(df[['Xcircle_X1', 'Xcircle_X2']])
```

### How many Clusters?

```python
import numpy as np
```

```python
np.unique(db.labels_)
```

```python
df['DB_Cluster'] = db.labels_
```

Visualize Clusters

```python
sns.scatterplot(x='Xcircle_X1', y='Xcircle_X2', hue='DB_Cluster', data=df, palette='spring')
```
