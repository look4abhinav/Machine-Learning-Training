### Load tensorflow

```python
import tensorflow as tf
#tf.enable_eager_execution()
```

```python
from sklearn.preprocessing import Normalizer
```

### Collect Data

```python
(train_x, train_y),(test_x, test_y) = tf.keras.datasets.boston_housing.load_data(test_split=0.2)
```

```python
train_x = train_x.astype('float32')
train_y = train_y.astype('float32')
test_x = test_x.astype('float32')
test_y = test_y.astype('float32')
```

```python
transformer = Normalizer()
train_x = transformer.fit_transform(train_x)
test_x = transformer.fit_transform(test_x)
```

### Building the graph in Keras

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Reshape((13,),input_shape=(13,)))

model.add(tf.keras.layers.Dense(100,activation='relu'))

model.add(tf.keras.layers.Dense(10,activation='relu'))

model.add(tf.keras.layers.Dense(1))
model.compile(optimizer='sgd', loss='mse')
```

```python
model.summary()
```

### Execute the Graph

```python
model.fit(train_x, train_y, epochs=100)
```

```python
model.get_weights()
```

### Save the model

```python
model.save('boston_housing_lr.h5')
```
